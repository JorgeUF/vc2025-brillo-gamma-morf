```python
def change_min_max(img, min_val, max_val):
    img2 = ((max_val - min_val) / 255) * img  + min_val    
    return img2.astype(np.uint8)
```

```python
img2 = change_min_max(img, 128, 255)
plt.figure(figsize=(10, 5))
plt.subplot(221)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(222)
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(223)
_ = plt.hist(img.flatten(), bins=256, range=(0, 255), color='gray', alpha=0.7)
plt.subplot(224)
_ = plt.hist(img2.flatten(), bins=256, range=(0, 255), color='gray', alpha=0.7)
```

```python
def gamma_correction(img, gamma):
    img2 = img / 255
    img2 = np.power(img2, gamma)
    img2 = (img2 * 255).astype(np.uint8)
    return img2
```

```python
img2 = gamma_correction(img, 3)
plt.figure(figsize=(10, 5))
plt.subplot(221)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(222)
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.subplot(223)
_ = plt.hist(img.flatten(), bins=256, range=(0, 255), color='gray', alpha=0.7)
plt.subplot(224)
_ = plt.hist(img2.flatten(), bins=256, range=(0, 255), color='gray', alpha=0.7)
```

# Apertura (dilation) y cierre (erosion)

```python
import cv2
img = cv2.imread('water_coins.jpg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
imgbin = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]
plt.figure(figsize=(10, 5))
plt.imshow(imgbin, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
imgbin.min(), imgbin.max()
```

```python
imgbin2 = 255 - imgbin
plt.figure(figsize=(10, 5))
plt.imshow(imgbin2, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
import numpy as np
```

```python
imgbin2_erosion = cv2.erode(imgbin2, np.ones((5, 5), np.uint8), iterations=4)
plt.figure(figsize=(10, 5))
plt.imshow(imgbin2_erosion, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
```

```python
imgbin2_dilation = cv2.dilate(imgbin2_erosion, np.ones((5, 5), np.uint8), iterations=4)
plt.figure(figsize=(10, 5))
plt.imshow(imgbin2_dilation, cmap='gray', vmin=0, vmax=255)
```

```python
_, markers = cv2.connectedComponents(imgbin2_erosion)
```

```python
np.unique(markers)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(markers, cmap='nipy_spectral', vmin=0, vmax=24)
plt.axis('off')
```
